```python
import requests

API_KEY = "AIzaSyCfgZM7gXT8fiA0YN2dbFZBJcsAuCiFI_o"
API_URL = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-2.0-flash:generateContent?key={API_KEY}"
PROMPT_FILE = "Prompt.txt"  
RAW_TEXT_FILE = "raw_data/document_to_anonymize.txt"
OUTPUT_FILE = "output.txt"
TEMPERATURE = 0.0

def generate_content(prompt_text: str, temperature: float) -> dict:
    """Envia o prompt para a API Gemini e retorna a resposta."""
    headers = {
        "Content-Type": "application/json"
    }
    body = {
        "contents": [
            {
                "parts": [
                    {"text": prompt_text}
                ]
            }
        ],
        "generationConfig": {
            "temperature": temperature
        }
    }
    response = requests.post(API_URL, headers=headers, json=body)
    return response.json()

try:
    with open(PROMPT_FILE, "r", encoding="utf-8") as f:
        prompt_with_examples = f.read()
except FileNotFoundError:
    print(f"Erro: {PROMPT_FILE} não encontrado.")
    exit()

try:
    with open(RAW_TEXT_FILE, "r", encoding="utf-8") as f:
        raw_text = f.read()
except FileNotFoundError:
    print(f"Erro: {RAW_TEXT_FILE} não encontrado.")
    exit()

final_prompt = f"{prompt_with_examples}\n\nRaw Text to Anonymize:\n{raw_text}"

output = generate_content(final_prompt, TEMPERATURE)

try:
    response_text = output['candidates'][0]['content']['parts'][0]['text']

    with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
        f.write(response_text)
    print(response_text)
except (KeyError, IndexError):
    print("Erro: Formato de resposta inesperado da API.")
    print(output)

```
